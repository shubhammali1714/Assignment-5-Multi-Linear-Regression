```python
#There Are 2 Method For Model Validation
# 1. Collinearity Check 
#2. Residual Analysis
# 1) Collinearity Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_r=smf.ols("RDS~ADS+MKTS",data=data1).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADS~RDS+MKTS",data=data1).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MKTS~RDS+ADS",data=data1).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
d1={'Variables':['RDS','ADS','MKTS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

 None variable has VIF>20, No Collinearity, so consider all varaibles in Regression equation

```python
# 2) Residual Analysis
# Test for Normality of Residuals (Q-Q Plot) use residual model.
sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
# Test for Homoscedasticity.
def standard_values(vals) : return (vals-vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standaradized residual values')
plt.show()
```

```python
# Test for errors or Residuals Vs Regressors.
# exog = x-variable & endog = y-variable
fig=plt.figure(figsize=(10,10))
sm.graphics.plot_regress_exog(model,'RDS',fig=fig)
plt.show()
```

#checking Outliers or Influencers

 There Are Two Method
 
1. Cook's Distance

2. Leverage value

```python
# 1. Cook's Distance.
#If Cook's distance > 1, then it's an outlier.
(c,_)=model.get_influence().cooks_distance
c
```

```python
# Plot the influence using the stem plot
fig=plt.figure(figsize=(15,10))
plt.stem(np.arange(len(data1)),np.round(c,5))
plt.xlabel("Row Index")
plt.ylabel("cooks Distance")
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

```python
# 2. Leverage Value.
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```
